# import matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# to shows us the figures in jupyter notebook
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
# we have 2 ways of plotting in this library 1- functional method 2- object oriented method
#1- Functional method
plt.plot(x,y)
# if we arenot in the jupyter notebook this fiigure will not appear unless we write plt.show()
# to get a label for x and y and a title for the graph we use the same method as matlab
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# to create multiple plots an the same figure we use subplot(no. of rows,no. of columns, the square that it will draw this plot in)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#2- object oriented method 
# the idea of this method is to use some special methods to plot the figure 
fig = plt.figure()
# the numbers in the list in the following method is used to determine the size of the square of figure
# add_axes([left,bottom,width,height of the axes]) 
# left: to move it from left to right
# bottom : to move it from bottom to up
# width : to control the width of figure
# height : to control the length of the height of the figure
# these numbers are from 0 to 1
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
# to make subplots
# function subplots(): it makes the figure and put axes on it based on the number of rows and cols we want 
# but function figure() : it just makes the figure so here we should add axes by ourselves by add_axes function
fig,axes = plt.subplots(nrows=3 , ncols=3)
```

```python
#to solve the overlabbing 
fig,axes = plt.subplots(nrows=3 , ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1 , ncols=2)

# axes is an array of list of axes i.e. we can iletrate over them
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
# we can alse index the axes 
fig,axes = plt.subplots(nrows=1 , ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

## figure size and DPI (dots per inches)

```python
# we put the size we want in a tuble (width , height)
fig = plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2 , ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
```

```python
# legend is a box that is used to denote which graph is belong to what 
# note that we incresed an attribute in plot method called label which indicate the name of this graph in the legend
# loc=10 it means to put the legend box in the center 
fig=plt.figure(figsize=(4,4))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x square')

ax.plot(x,x**3,label='x cube')

ax.legend(loc=10)
```

```python
# loc=0 it means to put the legend box in the best place 

fig=plt.figure(figsize=(4,4))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x square')

ax.plot(x,x**3,label='x cube')

ax.legend(loc=0)
```

```python
# loc=(a,b) it means to put the legend box in the height of a and width of b  

fig=plt.figure(figsize=(4,4))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x square')

ax.plot(x,x**3,label='x cube')

ax.legend(loc=(0.2,0.2))
```

# Plot Appearance

```python
# we can change the colour of the line bo color parameter 
# color parameter is taking a string with the basic colours in it like 'red' or taking the hex code of the line like '#FFFFFF' 
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple')
```

```python
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#000000')
```

```python

# linewidth is a parameter that we can control the width of the line by 
# its default = 1 
# if we want it to be thicker we increase it above 1 and if we want it to be thinner we decrease it below 1
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,linewidth = 1)
```

```python
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,linewidth = 20)
        
```

```python
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,linewidth = 0.5)
```

```python
# alpha parameter helps us to control the transparency of the line 
# we can control the color of line by this parameter 
# note line width can be written as lw 
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , alpha=0.5)
```

```python
# linestyle : is a parameter that can be used to change the style of line 
# we can make it '--' , ':' , '-.' , 'steps'
# we can write it as ls
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '--')
```

```python
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '-.')
```

```python
# we use marker parameter to put marks on the ordered pair (x,y) (single point)
# we use markersize to identify the size of this marker
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '-',marker = '+' , markersize = 20) 
```

```python
# markerfacecolor parameter is to change the color of the mark from the above
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '-',marker = 'o' , markersize = 20,
       markerfacecolor='red') 
```

```python
# markeredgewidth parameter is to change the size of the outside of the mark
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '-',marker = 'o' , markersize = 20,
       markerfacecolor='red',markeredgewidth=6) 
```

```python
# markeredgecolor parameter is to change the color of the mark from outside
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '-',marker = 'o' , markersize = 20,
       markerfacecolor='red',markeredgewidth=6,markeredgecolor='black')
```

```python
# if we want to zoom in the figure in specific numbers in x and y axes we use xlim & ylim methods 
# xlim & ylim take a list of 2 numbers : the start, the end point 
fig = plt.figure()
    
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple' ,lw = 3 , linestyle = '--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
