## Seaborn Library
#### seaborn is a statistical plotting library
#### it is built on the top of matplotlib
#### it has beautiful default styles
#### it works will with Pandas dataframe objects
#### link to seaborn documentation http://seaborn.pydata.org/

```python
# importing seaborn 
# the default we import it as sns
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# to keep visualizations appear in notebook 
%matplotlib inline
```

```python
# it is a built-in dataset 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# distplot gives us a histogram of the col. that we are passing it 
# kde is the line that pass on the graph ..... the default of it = True but here we make it False bec we don't want it now
# bins ....> it is the number of parts that the graph is parted into
sns.distplot(tips['total_bill'],kde = False , bins = 40)
```

```python
#jointplot is a method which give us a graph which denote 2 variables with each other
# (x..> it is the col. from the dataset that it will put it on x-axis ,y..> it is the col. from the dataset that it will put it on y-axis ,data is the name of our dataset )
# the default shape of the gragh is the scattered dots
sns.jointplot(x='total_bill' , y= 'tip',data=tips)
```

```python
# here we put kind parameter
# kind parameter can control the shape of graph 
# here the shape is hexagon 
sns.jointplot(x='total_bill' , y= 'tip',data=tips ,kind = 'hex')
```

```python
# here we changed the kind to reg 
# reg means linear regression
sns.jointplot(x='total_bill' , y= 'tip',data=tips ,kind = 'reg')
```

```python
#here we changed kind into kde 
# kde makes some circles around each others
sns.jointplot(x='total_bill' , y= 'tip',data=tips ,kind = 'kde')
```

##### pairplot()

```python
# pair plot method is making graphes to each 2 numerical variables
# it makes a histogram for each variable and itself
sns.pairplot(tips)
```

```python
# hue parameter is used when we want to know the effect of a categorical column in a graph 
# categorical ...> means a column has 2 string possible values  
sns.pairplot(tips , hue = 'sex')
```

```python
# palette helps us to change the degree of color of the graph 
sns.pairplot(tips , hue = 'sex' , palette='coolwarm')
```

#### rugplot()

```python
# rugplot like distplot takes one column and represent it 
# it represents this column as dashes 
# the difference from dashes and histogram is that in histogram it takes the number of dashes that indicate a number and represent the number of dashes as a bar
# this means that the numbers of dashes representing 10 equals 45 dashes 
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
# kde it is the sum of all normal distributions around a rugplot
#kde...> kernel density estimation 
# kdeplot to get kde only
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots
### barplot()

```python
# barplot make relation between a categorical value and a numerical value
# barplot takes x parameter that indicates the categorical column and y parameter which indicates the numerical column
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# esimator is a parameter that tell the barplot on what does it estimates its grapth
# the default estimator is mean 
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips , estimator= np.sum)
```

```python
# countplot is the same as barplot but y axis represents the number of occurance of x-axis
sns.countplot(x='sex',data=tips)
```

```python
# boxplot is used to represent relation between categorical and numerical value
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# we use hue parameter to be able to make another representation based on another categorical column 
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
# violin plots graph in the shape of violin  
# violin plots show more information than box plots
sns.violinplot(x='day', y='total_bill',data=tips)
```

```python
# in violinplot there is also hue parameter 
sns.violinplot(x='day', y='total_bill',data=tips, hue= 'sex')
```

```python
# split parameter makes the graph more obivous and easy to bring information from it
sns.violinplot(x='day', y='total_bill',data=tips, hue= 'sex', split=True)
```

```python
# stripplot is a scattered plot but with one categorical variable and the second is numerical variable
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# jitter parameter help us to see the density a little better 
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
# here we can also use hue
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
# there is a parameter which is called split which make it able to split the dots based on hue value
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split= True)
```

```python
# it is a combination of violin plot and strib plot 
# this function is not the better choice for large datasets because the points will be very far away from each others 
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# we can use violinplot and swarmplot together to overcome the problem of large dataset
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# factorplot can handle (show) any kind of plots based on the kind parameter 
# it is perferred to use for each graph its specific function
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

### grids

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
# pairplot is a simple specified type of pairGrid
sns.pairplot(iris)
```

```python
# it makes some grids without map anything on it 
sns.PairGrid(iris)
```

```python
# to get mapping using scatter 
# note: scatter is coming fron matplotlib library
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# we can specify some shapes to some grids 
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# FacetGrid
# is used to differentiate between some categorical columns
# it takes the dataset , the column which will have the same value in columns , a column which will have the same values in rows
# then we make map to our grid by any type we want
g=sns.FacetGrid(data= tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
# scatter needs 2 argument but distplot needs one
g=sns.FacetGrid(data= tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

# Regression plots

```python
# it gives us the most fitting line to the data 
sns.lmplot(x = 'total_bill' , y = 'tip' , data = tips)
```

```python
# here we can get a graph with 2 colours (one is male and the other is female)
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue='sex')
```

```python
# to specify the shapes of marker based on hue parammeter
# markers parameter is used
# this markers take a list of letters that are specified in the documentation
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue='sex',markers=['o','v'])
```

```python
# to make the markers biggers we use scatter_kws parameter
#this parameter takes a dictionary 
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
# we can make differentiation based on the sex col.
# i mean that to consider the left column for male and the right column for female
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,col='sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,col='sex',row='time')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,col='day',hue='sex',aspect=0.6,size=8)
```

# Style & color

```python
# to change the background color of the graph 
# the background is white
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
# the background is white with grids
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
# the background is dark(grey)
sns.set_style('dark')
sns.countplot(x='sex',data=tips)
```

```python
# the background is dark(grey) with grids
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# the background is white with ticks which are the borders 

sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
# we use function despine to remove the borders by default it remove the top and the right borders
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
# to change the siz of the graph we can use the figsize function from matplotlib library
plt.figure(figsize=(12,4))
sns.countplot(x='sex',data=tips)
```

```python
# to change the font size and the size and aspects of figure we use set_context function
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook',font_scale=3)
sns.countplot(x='sex',data=tips)
```
