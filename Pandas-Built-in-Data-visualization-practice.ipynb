```python
# i can use the graph on the data in dataframes
# we have 3 ways to plot
#1- is to use the function of the graph
df1['A'].hist(bins=50)
```

```python
#2- is to use plot functionand inside it we give it the kind of graph we want
df1['A'].plot(kind = 'hist')
```

```python
# we can plot one column or the whole dataset
# 3- we can use plot function . the function of the graph we want
df1['A'].plot.hist()

```

```python
df2.plot.area()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
# plot as line 
df1.plot.line(x='A',y='B')
```

```python
# can take any parameter from the matplotlib library such as figsize
df1.plot.line(x='A',y='B',figsize=(12,3))
```

```python
# plot as scatter
df1.plot.scatter(x='A',y='B')
```

```python
# plot as scatter 
# plot 3 variables with each other
# we represent c by the color
df1.plot.scatter(x='A',y='B',c='C')
```

```python
# c for color : it is possible to pass a string with the name of the color 
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
# here the location of point indicates a and b and the size of the point indicates c value
# s for size : it is possible to pass a number
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

```python
# plot as box plot
df2.plot.box()
```

```python
df= pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
# hexbin like scatter 
# it differs in the hexagon shape of each point
df.plot.hexbin(x='a',y='b' , gridsize=25)
```

```python
# to plot kde (estimator)
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```
